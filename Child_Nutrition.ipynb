# Child Nutrition Calculator
### The calculator should be able to store the details of the child such as the name, age, gender, height and weight.

```python
name = input("Enter name of child: ")
gender = input("Enter the gender: ")
age = int(input("Enter age of child: "))
height = int(input("Enter height in inches: "))
weight = int(input("Enter weight in pounds: "))
```

## Display the BMI(Body Mass Index) based on Height & Weight:


```python
BMI = int(weight/(height**2)*703)
if BMI < 16:
    print("Severely Underweight")
elif 16 >= BMI < 18.5:
    print("Underweight")
elif 18.5 >= BMI < 25:
    print("Healthy")
elif 25 >= BMI < 30:
    print("Overweight")
else:
    print("Obese")
print("BMI of ",name,"is: ",BMI)
```

```python
print("Provide parameter as per grams for calorie consumption")
milk1 = int(input("Milk: "))
veg1 = int(input("Vegetables: "))
meat1 = int(input("Meat: "))
lentis1 = int(input("lentis: "))
egg1 = int(input("Egg: "))
rice1 = int(input("Rice: "))
```

```python
Milk = 100 / 100 * milk1
Egg = 155 / 100 * egg1
Rice = 130 / 100 * rice1
Lentis = 113 / 100 * lentis1
Vegetable = 85 / 100 * veg1
Meat = 143 / 100 * meat1
total = Milk + Egg + Rice + Lentis + Vegetable + Meat
```

## Display the minimum daily calorie requirement based on age: 


```python
print(f"The Daily Calorie Consumption Of {name} is {total} And (He/She) is: ")
if age > 0 and age <= 2:
    total >= 800
    print("Nourished")
elif age >= 2 and age <= 4:
    total >= 1400
    print("Nourished")
elif age >= 4 and age <= 8:
    total >= 1800
    print("Nourished")
else:
    print("Not Nourished")
```
